```python
from sklearn.preprocessing import StandardScaler
X = df[['age','annual_income','spending_score']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(X_scaled[:5])
```

```python
from sklearn.cluster import KMeans
inertia = []
for k in range (1,11):
  kmeans  = KMeans(n_clusters =k, random_state = 42)
  kmeans.fit(X_scaled)
  inertia.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), inertia, marker='o')
plt.title("Elbow Method for Optimal k")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, random_state = 42)
```

```python
df['Cluster'] = kmeans.fit_predict(X_scaled)
print(df.head())
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x = 'annual_income',y='spending_score', hue = 'Cluster', palette ='Set2', s=100, alpha=0.6)
plt.title("Customer Segment (K-means Clustering)")
plt.show()
```

```python
centroids = kmeans.cluster_centers_
print("Cluster Centers (Centroids):\n", scaler.inverse_transform(centroids))
```
